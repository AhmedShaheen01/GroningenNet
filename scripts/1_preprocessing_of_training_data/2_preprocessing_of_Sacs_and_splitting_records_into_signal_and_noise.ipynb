### This script preprocess sac files sorted in the events folders, then sort them  into noise and signal folders.

The tree of folders as follows: -Event
                                    |__ Noise
                                    |__ Signal
                                    |__ Trash

```python
def unify_start_time(st):
    st.sort(['starttime'])       #sort function works in place
    start_time=st[-1].stats['starttime']
    st.trim(start_time)
    st.sort(['station'])
```

```python
def unify_end_time(st):
    st.sort(['endtime'])
    end_time=st[0].stats['endtime']
    st.trim(endtime=end_time)
    st.sort(['station'])
```

```python
unify_start_time(st2)
unify_end_time(st2)
```

```python
st_signal=st2.copy()
st_noise=st2.copy()
```

```python
def extract_noise(st):
    st.trim(st[0].stats['starttime'],st[0].stats['starttime']+90)
    for tr in st:
        tr.stats.sac['nzyear']=tr.stats.starttime.year
        tr.stats.sac['nzjday']=tr.stats.starttime.julday
        tr.stats.sac['nzhour']=tr.stats.starttime.hour
        tr.stats.sac['nzmin']=tr.stats.starttime.minute
        tr.stats.sac['nzsec']=tr.stats.starttime.second
        tr.stats.sac['nzmsec']=tr.stats.starttime.microsecond
        #tr.stats.sac['npts']=9000
```

```python
def extract_signal(st):
    st.trim(st[0].stats['endtime']-90,st[0].stats['endtime'])
    for tr in st:
        tr.stats.sac['nzyear']=tr.stats.starttime.year
        tr.stats.sac['nzjday']=tr.stats.starttime.julday
        tr.stats.sac['nzhour']=tr.stats.starttime.hour
        tr.stats.sac['nzmin']=tr.stats.starttime.minute
        tr.stats.sac['nzsec']=tr.stats.starttime.second
        tr.stats.sac['nzmsec']=tr.stats.starttime.microsecond
        #tr.stats.sac['npts']=9000
```

```python
extract_noise(st_noise)
extract_signal(st_signal)
```

```python
def create_folder(directory):
    try:
        if not os.path.exists(directory):
            os.makedirs(os.getcwd()+'/'+directory)
    except OSError:
        print('Error: Creating directory. '+directory)
```
